```python
import pandas as pd 
import numpy as np
```

```python
# Description of the data can be found in the `data/regression` folder
df = pd.read_csv('./data/regression/OLSChallenge.csv', encoding="ISO-8859-1")
df.head()
```

```python
df.columns
```

```python
# Calculate the VIF Score of 
# incident rate against 
# the feature_set 
# 'medIncome', 'MedianAge', 'povertyPercent'

subset = df[['incidenceRate', 'medIncome', 'MedianAge', 'povertyPercent']]
subset.head()
```

```python
import statsmodels.api as sm
```

```python
X = subset.drop(columns=['incidenceRate'])
y = subset['incidenceRate']

ols = sm.OLS(y, X).fit()
display(ols.summary())
r2 = ols.rsquared 
vif = 1/(1-r2)
vif
```
