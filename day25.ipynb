2. A sampling distritubion is the distribution of sample means. If you take the mean of a sample from a population repeatedly, these means will form a normal distrubtion.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# code taken from https://jakevdp.github.io/PythonDataScienceHandbook/05.13-kernel-density-estimation.html
def make_data(N, f=0.3, rseed=1):
    rand = np.random.RandomState(rseed)
    x = rand.randn(N)
    x[int(f * N):] += 5
    return x


def plot_dist(X, name="X"):
    plt.figure(figsize=(8, 5))
    sns.distplot(X, bins=30)
    plt.xlabel("X")
    plt.ylabel("PDF")
    plt.title(f"PDF of {name}")
    plt.show()
```

```python
# generating bimodal distribution
X = make_data(1000)
plot_dist(X)
```

```python
# take a bootstrap sample of X
# a bootstrap sample (X_bootstrap) of X has the following properties
# X has 1000 points in it. X_bootstrap will also have 1000 points in it. 
# Each point in X_bootstrap will come from X randomly with replacement
# You can use np.random.choice to conduct this sampling

X_bootstrap = np.random.choice(X, size=1000, replace=True)
# plot X_bootstrap
plot_dist(X_bootstrap, name="X_bootstrap")
```

* Notice this has the same shape as X, but the data tends 
to be more centered around the values with a higher probablity
* Notice the y-axis has higher values in the bootstrap sample plot
    * this doesn't always happen, but it tends to happen

```python
sample_means = []

# create a loop of 30 iterations that takes bootstrap samples and takes the mean of the sample
# append this mean to sample_means

for i in range(30):
    sample_means.append(np.random.choice(X, size=1000, replace=True).mean())

sample_means
```

```python
sns.distplot(sample_means)
```
