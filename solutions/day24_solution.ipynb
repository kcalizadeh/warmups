# Section A
There was a recent poll taken to measure the salaries of tenants in a cluster of apartment complexes.  After some statistical analysis the outcome was that the mean of the sample was \\$65k and the sample standard deviation was \\$4k. 

Given this information:

What is the probability that a tenant, chosen at random, will

* Have a salary over \\$75k
* Have a salary less than \\$40k
* Have a salary between \\$60k and \\$72k

```python
# code here
import scipy.stats as scs # import scipy.stats as stats
import numpy as np 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# over 75k
dist = scs.norm(loc=65, scale=4) # loc = mean, scale = std
1 - dist.cdf(75)
1 - scs.norm.cdf(75, 65, 4)
```

```python
dist.cdf(40)
```

```python
dist.cdf(72)-dist.cdf(60)
```

# Section B
* plot the pdf of the distribution of salaries
* plot the cdf of the distribution of salaries

```python
# ppf = percent point function
x_axis = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 1000) # generate our xaxis
probs = dist.pdf(x_axis) # generate our yaxis with our xaxis
```

```python
plt.plot(x_axis, probs)
plt.title("PDF of our Dist")
plt.ylabel("PDF")
plt.show()
```

```python
x_axis
cdfs = dist.cdf(x_axis)
```

```python
plt.plot(x_axis, cdfs)
plt.show()
```

```python
x_axis = np.linspace(0, 1, 1000)
ppfs = dist.ppf(x_axis)
```

```python
plt.plot(x_axis, ppfs)
plt.title("PPF of our Dist")
plt.ylabel("PPF")
plt.show()

# mapping of percentile
# where is the 30th percentile
# inverse mapping of cdf
```
