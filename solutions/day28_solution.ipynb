```python
import numpy as np 

import scipy.stats as scs
from statsmodels.stats.power import tt_ind_solve_power

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load in arrays for the following exercises

arr1 = np.loadtxt("../data/norm_dist1.txt")
arr2 = np.loadtxt("../data/norm_dist2.txt")
```

```python
# calculate Cohen's d (effect_size) for the given sample distributions
def cohens_d(arr1, arr2):
    n_arr1 = arr1.shape[0]
    n_arr2 = arr2.shape[0]
    ddof = n_arr1 + n_arr2 - 2
    numerator = np.abs(np.mean(arr1) - np.mean(arr2))
    denominator = np.sqrt((n_arr1*np.std(arr1, ddof=1)**2 + n_arr2*np.std(arr2, ddof=1)**2)/ddof)
    return numerator / denominator

effect_size = cohens_d(arr1, arr2)
print(effect_size)
```

```python
# Use the effect size to calculate the power score with an alpha of 0.10
effect_size = effect_size
alpha = 0.10
nobs1 = arr1.shape[0]

power = tt_ind_solve_power(effect_size=effect_size, nobs1=nobs1, alpha=alpha)
print(power)
```

```python
def plot_distributions(arr1, arr2, title1='arr1', title2='arr2', bins=20):
    sns.distplot(arr1, label=title1, bins=bins)
    sns.distplot(arr2, label=title2, bins=bins)
    plt.legend()
    plt.show()
```
